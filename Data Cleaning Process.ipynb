### Import Libraries and Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('messy_customer_sales_data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe().round()
```

### Check Missing Values

```python
data.isnull().sum()
```

#### Remove Duplicate Records

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data.head()
```

#### Clean Customer ID

```python
data['Customer_ID'].isnull().sum()
```

```python
# fill missing values
data['Customer_ID'] = data['Customer_ID'].fillna('Unknown')
```

#### Clean Gender Column

```python
print(data['Gender'].unique())
```

```python
data['Gender'] = data['Gender'].str.strip().str.upper()   #convert into uppercase
```

```python
print(data['Gender'].head(5))
```

```python
# replace values
data['Gender'] = data['Gender'].replace({
    'M':'Male',
    'F':'Female',
    'MALE':'Male',
    'FEMALE':'Female'
})
```

```python
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

#### Clean Age Column

```python
data['Age'].head(5)
```

```python
data['Age'] = data['Age'].astype(str)
data['Age'] = data['Age'].str.extract('(\d+)')
```

```python
data['Age'] = pd.to_numeric(data['Age'])
```

```python
# fill missing ages
data['Age'] = data['Age'].fillna(data['Age'].median())
```

#### Clean City Column

```python
data['City'].unique()
```

```python
data['City'] = data['City'].str.strip().str.title()  # strip removes extra space
```

```python
data['City'] = data['City'].fillna(data['City'].mode()[0])
```

#### Clean Country Column

```python
data['Country'].unique()
```

```python
data['Country'] = data['Country'].str.strip().str.upper()
```

```python
data['Country'] = data['Country'].replace({
    'IND':'India'
})
```

```python
data['Country'] = data['Country'].fillna('India')
```

#### Convert Date Columns

```python
data['Signup_Date'] = pd.to_datetime(data['Signup_Date'])
```

```python
data['Last_Purchase_Date'] = pd.to_datetime(
    data['Last_Purchase_Date'])
```

```python
data['Last_Purchase_Date'] = data['Last_Purchase_Date'].fillna(
    data['Last_Purchase_Date'].median()
)
```

#### Clean Purchase Amount

```python
data['Purchase_Amount'].isnull().sum()
```

```python
data['Purchase_Amount'] = data['Purchase_Amount'].fillna(
    data['Purchase_Amount'].median()
)
```

```python
data['Purchase_Amount'].isnull().sum()
```

#### Clean Feedback Score

```python
data['Feedback_Score'] = data['Feedback_Score'].fillna(
    data['Feedback_Score'].median()
)
```

```python
data['Feedback_Score'].isnull().sum()
```

```python
data['Feedback_Score'] = data['Feedback_Score'].astype(int)
```

### Verify Cleaned Data

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
print(data.head())
```

#### Save Cleaned Dataset

```python
data.to_csv("cleaned_customer_sales_data.csv", index=False)

print("Dataset Saved Successfully")
```
